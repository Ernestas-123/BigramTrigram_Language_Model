```python
import copy          
import numpy as np 

# Make sure file in the same directory
with open("sentences_short.txt", "r") as f:
    sentences_short = [i[len("<s>"):-len("</s>") - 1].lower().strip() for i in f]
    

# Each sentence is a list (List of lists)
words_long = [w.split(" ") for w in sentences_short]

for i in words_long:
    while "" in i:
        i.remove('')

# Each word it's own element in list (List of words)
words_list = []
for s in words_long:
    for i in range(len(s)):
        words_list.append(s[i])
        
# Unique words
unique_words = set(words_list)
```

#### Task 1: Count Word Frequencies

```python
# Counts how many times a unique word appears in the list of words, becomes a dictionary
word_frequencies = {}
for i in unique_words:
    word_frequencies[i] = words_list.count(i)
    
# Sort
word_frequencies = dict(sorted(word_frequencies.items()))
word_frequencies
```

#### Task 2: Find the Most Frequent Word

```python
values_list = []  # Store occurences of each unique word
for i in word_frequencies.values():  # Access values, not key
    values_list.append(i)
max(values_list)  # The amount of times the most common word occured
```

```python
most_frequent = {}  # Maybe more than 1

# Iterate through the dictionary, find the keys with corresponding max value
for i in word_frequencies:  
    if word_frequencies[i] == max(values_list):
        most_frequent[i] = max(values_list)
        
most_frequent
```

#### Task 3: Convert Word Frequencies to a Probability Distribution

```python
word_distribution = copy.deepcopy(word_frequencies)  # Copies original, when first created

# Total amount of words
words_length = len(words_list)

# In dictionary for the values, replace occurences of a word with it's probability distribution
for key, value in word_distribution.items():
    word_distribution[key] = value / words_length
    
word_distribution
```

#### Task 15: Build a Language Model Object

```python
import copy
import numpy as np
from collections import defaultdict
class trigram_language_model:

    def __init__(self):
        self.conditional_distribution = None
        
    def train(self, corpus_file_name):
        self.corpus_file_name = corpus_file_name  # Ensure this is a string
        self.conditional_frequencies = {}
        self.trigrams = []
        
        with open(self.corpus_file_name, "r") as f:
            self.sentences_short = [i.lower().strip() for i in f]
            
        self.words_long = [w.split(" ") for w in self.sentences_short]

        for i in self.words_long:  # Double space before </s>
            while "" in i:
                i.remove('')

        self.words_list = []  # List of all words in file
        for s in self.words_long:
            for i in range(len(s)):
                self.words_list.append(s[i])
                
        self.unique_words = set(self.words_list)  # Unique Words
        
        # Frequency of each word
        self.word_frequencies = {}
        for i in self.unique_words:
            self.word_frequencies[i] = self.words_list.count(i)
            
        # Unigram
        self.word_distribution = copy.deepcopy(self.word_frequencies)
        self.words_length = len(self.words_list)
        for key, value in self.word_distribution.items():
            self.word_distribution[key] = value / self.words_length
        
        # Trigram
        self.trigrams = []

        for i in self.words_long:
            self.length = len(i)  # How many words in sentence

            for j in range(self.length):
                self.trigrams.append(i[j:j+3])

        for i in self.trigrams:
            if len(i) == 1:
                self.trigrams.remove(i)

        for i in self.trigrams:
            if len(i) == 2:
                self.trigrams.remove(i)
        
        self.unique_trigrams = set(tuple(trigram) for trigram in self.trigrams)
        self.trigram_frequencies = {}

        for i in self.unique_trigrams:
            self.trigram_frequencies[i] = self.trigrams.count(list(i))

        self.trigram_frequencies = dict(sorted(self.trigram_frequencies.items()))
        
        
        # bigrams
        self.bigrams = []

        for i in self.words_long:
            self.length = len(i)  # How many words in sentence

            for j in range(self.length):
                self.bigrams.append(i[j:j+2])

        for i in self.bigrams:
            if len(i) == 1:
                self.bigrams.remove(i)

        self.unique_bigrams = set(tuple(bigram) for bigram in self.bigrams)
        
        self.bigram_frequencies = {}
        for i in self.unique_bigrams:
            self.bigram_frequencies[i] = self.bigrams.count(list(i))
        
        # I need bigrams for trigrams
        self.conditional_frequencies = {}
        
        for i in list(self.unique_bigrams):
            self.conditional_frequencies[i] = {}

        for key in self.conditional_frequencies:
            self.conditional_frequencies[key] = self.corresponding_word_dic2(key[0:2])

        self.new_conditional_freq = defaultdict(dict) #Empty nested dictionary, defaultdict
        for outerkey, outervalue in self.conditional_frequencies.items():
            new_outerkey = list(outerkey)
            
            for innerkey, innervalue in self.conditional_frequencies[outerkey].items():
                new_innerkey = list(innerkey)
                
                is_sublist = all(element in new_innerkey for element in new_outerkey)
                if is_sublist:
                    for i in range (0, 1+1):
                        new_innerkey.remove(new_outerkey[i])
                        
                    self.new_conditional_freq[outerkey][tuple(new_innerkey)] = innervalue 

        # Sort
        self.new_conditional_freq = dict(sorted(self.new_conditional_freq.items())) 
        
        self.conditional_distribution = copy.deepcopy(self.new_conditional_freq)

        for outerkey, outervalue in self.conditional_distribution.items():
            sum1 = sum(self.conditional_distribution[outerkey].values())

            for innerkey, innervalue in self.conditional_distribution[outerkey].items():
                self.conditional_distribution[outerkey][innerkey] = innervalue / sum1

        # Sort
        self.conditional_distribution = dict(sorted(self.conditional_distribution.items()))
        
        message = f"The trigram model has been trained using the file '{self.corpus_file_name}'"
        print(message)
        
        
    def corresponding_word_dic2(self, first_2word):  # Make sure tuple with 2 strings
        self.first_2word = first_2word
        self.temp_dic = {}
        self.store = []
        
        for key, value in self.trigram_frequencies.items():
            if self.first_2word == key[0:2]:
                self.temp_dic[key] = value
                self.store.append(value)

        return self.temp_dic
    
    def generate_sentence(self, start_word):
        self.start_word = start_word
        self.sentence = ['<s>', self.start_word]
        
        while True:
            bigram_1 = tuple(self.sentence[-2::])
            keys_list = [key[0] for key in self.conditional_distribution[bigram_1].keys()]
            values_list = list(self.conditional_distribution[bigram_1].values())

            word = np.random.choice(keys_list, 1, p = np.array(values_list))[0]
            self.sentence.append(word)

            if word == "</s>":
                return " ".join(self.sentence)
        
        
```

*Note
Make sure input file has < s> at the start of each sentence and < /s> at the end of each sentence.
Each sentence must be on the same line, as the model reads line by line.
Ensure there are no special or different format characters that might interfere in the file.


*Note: When selecting the 'first word' to input, the word selected must have been at the start of a atleast 1 sentence in the file (used to train the model).
i.e in the file there are a lot of sentences in the form: < s> word1, word2, word3 etc < /s>.
Your word you input into the model must be like word1, AFTER < s>.
Otherwise you will receive a key error.

This can be fixed by modifying the model to first input a word, use bigram conditional probabilities, then switch to trigram conditional probabilities.
But even then depending on the size of your file you can receive a key error. i.e you use a word not used in the training file.

It took too long for my computer to load "sentences_long.txt". So online I found "Reservoir Sampling" which takes a sample of known size q from a population n, where n is unknown quantity.

This randomly samples a 0.1, 1%, 10%, 20% etc of the 1 million lines in "sentences_long.txt" and creates and stores them in a seperate .txt file called "sentences_sample.txt". With each line having 1/n chance to be selected.

This sample can be shown that the class trigram_language_model works with different .txt files of similar format.

### Samples at random without replacement lines from a .txt file

```python
import random

sample_percentage = 0.001

def sample_lines(input_file, output_file, sample_percentage):
    # Open file and store file object as variable name "input_file"
    with open(input_file, "r") as input_file, open(output_file, "w") as output_file:
        
        total_lines = sum(1 for line in input_file)
        
        num_lines_to_sample = int(total_lines * (sample_percentage))
        
        # Reset file pointer to the beginning of the input file
        input_file.seek(0)
        
        sampled_lines = []
        
        # Enumerate returns a tuple of the element and index number of an object
        for line_num, line in enumerate(input_file, start=1): # first, the sample lines same as 1st 200
            if len(sampled_lines) < num_lines_to_sample:
                sampled_lines.append(line)
                
            # Reservoir Sampling?
            else:
                replace_index = random.randint(0, line_num - 1)
                if replace_index < num_lines_to_sample: # always true
                    sampled_lines[replace_index] = line
                    
        output_file.writelines(sampled_lines)
                        
sample_lines("sentences_long.txt", "sentences_sample.txt", 0.001) # Sample of 0.1%
```
